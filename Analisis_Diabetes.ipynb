```python
import pandas as pd
import matplotlib.pyplot as plt

# ==============================
# 1. CARGA DEL DATASET
# ==============================
df = pd.read_csv("diabetes_012_health_indicators_BRFSS2015.csv")

# Vista inicial
print("Primeras filas del dataset:")
display(df.head())


```

```python
# ==============================
# 2. EXPLORACIÓN GENERAL
# ==============================
print("Información general del dataset:")
df.info()

print("\nDimensiones del dataset (filas, columnas):")
print(df.shape)

print("\nValores faltantes por columna:")
print(df.isnull().sum())

print("\nEstadísticas descriptivas generales:")
display(df.describe())
```

```python
# ==============================
# 3. SELECCIÓN DE VARIABLES
# ==============================
variables = ["Diabetes_012", "BMI", "Age", "PhysActivity", "Smoker"]
df_sel = df[variables].copy()

# Convertir algunas variables a enteros para mejor visualización
df_sel["Diabetes_012"] = df_sel["Diabetes_012"].astype(int)
df_sel["Age"] = df_sel["Age"].astype(int)
df_sel["PhysActivity"] = df_sel["PhysActivity"].astype(int)
df_sel["Smoker"] = df_sel["Smoker"].astype(int)

print("Primeras filas de las variables seleccionadas:")
display(df_sel.head())
```

```python
# ==============================
# 4. ETIQUETAS PARA VARIABLES CATEGÓRICAS
# ==============================
df_sel["Diabetes_012_label"] = df_sel["Diabetes_012"].map({
    0: "No diabetes",
    1: "Prediabetes",
    2: "Diabetes"
})

df_sel["PhysActivity_label"] = df_sel["PhysActivity"].map({
    0: "No realiza actividad física",
    1: "Realiza actividad física"
})

df_sel["Smoker_label"] = df_sel["Smoker"].map({
    0: "No fumador",
    1: "Fumador"
})
```

```python
# ==============================
# 5. ANÁLISIS UNIVARIADO - BMI
# ==============================
print("Estadísticas descriptivas de BMI:")
display(df_sel["BMI"].describe())

```

```python
# Histograma
plt.hist(df_sel["BMI"], bins=30, color="purple", edgecolor="black")
plt.title("Distribución del Índice de Masa Corporal (BMI)")
plt.xlabel("BMI")
plt.ylabel("Cantidad de personas")
plt.grid(axis="y", alpha=0.3)
plt.show()
```

# Entrega 2 – Preparación de los datos y análisis de relaciones

En esta etapa se realiza el tratamiento de los datos, la construcción del dataset final y el análisis de relaciones entre variables relevantes para el estudio de la diabetes.

## 1. Tratamiento de los datos

En esta sección se analizan aspectos relacionados con la calidad de los datos, incluyendo valores faltantes, detección de outliers y posibles transformaciones.

### 1.1. Valores faltantes

```python
faltantes = df_sel.isnull().sum()

print("Análisis de valores faltantes:\n")
print(faltantes)

if faltantes.sum() == 0:
    print("\n✔ No se detectan valores faltantes en las variables seleccionadas.")
    print("✔ No es necesario aplicar técnicas de imputación.")
else:
    print("\n⚠ Se detectaron valores faltantes.")
    print("⚠ Se procederá a aplicar técnicas de imputación.")
```

### 1.2 Detección de outliers

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.boxplot(x=df_sel["BMI"], color="#9370DB", fliersize=2)
plt.title("Detección de outliers en BMI")
plt.show()
```

En el gráfico se observan múltiples valores atípicos (outliers), representados como puntos fuera de los límites del boxplot.  
Estos corresponden a valores extremos del índice de masa corporal.

```python
Q1 = df_sel["BMI"].quantile(0.25)
Q3 = df_sel["BMI"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df_sel[(df_sel["BMI"] < lower_bound) | (df_sel["BMI"] > upper_bound)]

print("Límite inferior:", lower_bound)
print("Límite superior:", upper_bound)
print("Cantidad de outliers:", outliers.shape[0])
```

```python
outliers.head()
```

### Análisis de outliers en BMI

Los valores atípicos identificados corresponden principalmente a índices de masa corporal elevados, asociados a casos de obesidad severa.

El análisis de estos registros muestra que muchos de estos individuos no realizan actividad física, lo cual podría influir en el aumento del BMI. Sin embargo, se observan tanto casos con diabetes como sin ella, lo que indica que el BMI no es el único factor determinante.

Se concluye que estos valores representan casos reales del dataset, por lo que no fueron eliminados.

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df_sel["BMI"], color="#9370DB", showfliers=False)
plt.title("Distribución del BMI (sin visualización de outliers)")
plt.xlabel("BMI")
plt.show()
```

El boxplot de la variable BMI presenta una gran cantidad de valores atípicos. Para mejorar la visualización, se optó por no mostrarlos en el gráfico, aunque se reconoce su existencia en el dataset.

## 1.3 Transformaciones de los datos

En esta etapa se aplican transformaciones sobre el dataset con el objetivo de mejorar la calidad, interpretación y análisis de las variables.

Estas transformaciones no modifican el significado de los datos, sino que permiten representarlos de manera más clara y adecuada para su análisis.

### Renombrado de variables

```python
df_esp = df.rename(columns={
    "Diabetes_012": "Estado_Diabetes",
    "HighBP": "Presion_Alta",
    "HighChol": "Colesterol_Alto",
    "CholCheck": "Control_Colesterol",
    "BMI": "BMI",
    "Smoker": "Fumador",
    "Stroke": "ACV",
    "HeartDiseaseorAttack": "Enfermedad_Cardiaca",
    "PhysActivity": "Actividad_Fisica",
    "Fruits": "Consumo_Frutas",
    "Veggies": "Consumo_Verduras",
    "HvyAlcoholConsump": "Consumo_Alcohol_Alto",
    "AnyHealthcare": "Acceso_Salud",
    "NoDocbcCost": "Sin_Medico_por_Costo",
    "GenHlth": "Salud_General",
    "MentHlth": "Salud_Mental",
    "PhysHlth": "Salud_Fisica",
    "DiffWalk": "Dificultad_Caminar",
    "Sex": "Sexo",
    "Age": "Edad",
    "Education": "Educacion",
    "Income": "Ingreso"
})

df_esp.head()
```

Se renombraron las variables al idioma español con el objetivo de mejorar la claridad y facilitar la interpretación del análisis.

### Conversión de tipos de datos

```python
df_esp["Estado_Diabetes"] = df_esp["Estado_Diabetes"].astype(int)
df_esp["Edad"] = df_esp["Edad"].astype(int)
df_esp["Actividad_Fisica"] = df_esp["Actividad_Fisica"].astype(int)
df_esp["Fumador"] = df_esp["Fumador"].astype(int)
```

Algunas variables se encontraban en formato decimal (float), a pesar de representar categorías discretas.  
Por este motivo, se convirtieron a tipo entero para mejorar su interpretación sin alterar los datos originales.

### Generación de etiquetas descriptivas

```python
from IPython.display import display

df_esp["Estado_Diabetes_Descripcion"] = df_esp["Estado_Diabetes"].map({
    0: "No diabetes",
    1: "Prediabetes",
    2: "Diabetes"
})

df_esp["Actividad_Fisica_Descripcion"] = df_esp["Actividad_Fisica"].map({
    0: "No realiza actividad física",
    1: "Realiza actividad física"
})

df_esp["Fumador_Descripcion"] = df_esp["Fumador"].map({
    0: "No fumador",
    1: "Fumador"
})

display(df_esp[["Estado_Diabetes", "Estado_Diabetes_Descripcion"]].head())

display(df_esp[["Actividad_Fisica", "Actividad_Fisica_Descripcion"]].head())

display(df_esp[["Fumador", "Fumador_Descripcion"]].head())
```

Se generaron variables con etiquetas descriptivas para facilitar la interpretación de los resultados y mejorar la claridad de los gráficos.

### Clasificación del índice de masa corporal (BMI)

```python
def categorizar_bmi(valor):
    if valor < 18.5:
        return "Bajo peso"
    elif valor < 25:
        return "Normal"
    elif valor < 30:
        return "Sobrepeso"
    else:
        return "Obesidad"

df_esp["Categoria_BMI"] = df_esp["BMI"].apply(categorizar_bmi)

df_esp[["BMI", "Categoria_BMI"]].head()
```

Se creó una variable categórica a partir del BMI para facilitar el análisis, clasificando los valores en categorías estándar utilizadas en el ámbito de la salud.

## 2. Construcción del dataset final

A partir de las transformaciones realizadas, se construyó un dataset final que incluye variables relevantes para el análisis de la diabetes.

Se seleccionaron variables relacionadas con el estado de salud, hábitos de vida y condición física de los individuos.

```python
df_final = df_esp[[
    "Estado_Diabetes",
    "Estado_Diabetes_Descripcion",
    "BMI",
    "Categoria_BMI",
    "Edad",
    "Actividad_Fisica",
    "Actividad_Fisica_Descripcion",
    "Fumador",
    "Fumador_Descripcion"
]].copy()

df_final.head()
```

```python
df_final.info()
```

El dataset final está compuesto por 253.680 registros y 9 variables seleccionadas, las cuales fueron consideradas relevantes para el análisis del problema planteado.

No se detectaron valores faltantes en ninguna de las variables, lo que indica una alta calidad del dataset.

Asimismo, se observa una correcta asignación de tipos de datos:

Variables numéricas (como BMI)
Variables categóricas codificadas (como Estado_Diabetes)
Variables categóricas descriptivas (labels)

Esto permite realizar un análisis adecuado tanto desde el punto de vista estadístico como interpretativo.

## 3. Análisis bivariado

En esta sección se analizan las relaciones entre variables del dataset con el objetivo de identificar patrones asociados al estado de diabetes.

Se seleccionaron variables relevantes vinculadas a la condición física y hábitos de vida.

### 3.1 Relación entre BMI y estado de diabetes

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))

sns.boxplot(
    x="Estado_Diabetes_Descripcion",
    y="BMI",
    hue="Estado_Diabetes_Descripcion",
    data=df_final,
    palette="Purples",
    legend=False
)

plt.title("Relación entre BMI y estado de diabetes")
plt.xlabel("Estado de diabetes")
plt.ylabel("Índice de masa corporal (BMI)")

plt.show()
```

Interpretacion: El gráfico muestra la distribución del índice de masa corporal (BMI) en función del estado de diabetes.
Se observa que la mediana del BMI es más baja en individuos sin diabetes, mientras que aumenta en los grupos con prediabetes y diabetes. En particular, las personas con diabetes presentan valores más elevados de BMI en comparación con los otros grupos.
La dispersión de los datos indica que los valores de BMI tienden a ser más altos en individuos con prediabetes y diabetes, lo que sugiere una relación positiva entre el índice de masa corporal y la presencia de la enfermedad.
Estos resultados indican que un mayor BMI podría estar asociado a un mayor riesgo de desarrollar diabetes, lo cual es consistente con el conocimiento en el ámbito de la salud.

### 3.2 Relación entre actividad física y diabetes

```python
tabla_af = pd.crosstab(
    df_final["Actividad_Fisica_Descripcion"],
    df_final["Estado_Diabetes_Descripcion"],
    normalize="index"
)

tabla_af.plot(
    kind="bar",
    stacked=True,
    figsize=(7,5),
    color=["#6A0DAD", "#B497D6", "#3F007D"]  
)

plt.title("Proporción de diabetes según actividad física")
plt.xlabel("Actividad física")
plt.ylabel("Proporción")

plt.xticks(rotation=0)

plt.legend(title="Estado de diabetes", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

Se observa que la proporción de individuos con diabetes es mayor en el grupo que no realiza actividad física. Esto indica que el sedentarismo podría estar asociado a un mayor riesgo de desarrollar la enfermedad.

### 3.3 Relación entre tabaquismo y diabetes

```python
tabla_sm = pd.crosstab(
    df_final["Fumador_Descripcion"],
    df_final["Estado_Diabetes_Descripcion"],
    normalize="index"
)

tabla_sm.plot(
    kind="bar",
    stacked=True,
    color=["#6A0DAD", "#B497D6", "#3F007D"],
    figsize=(7,5)
)

plt.title("Proporción de diabetes según tabaquismo")
plt.xlabel("Tabaquismo")
plt.ylabel("Proporción")

plt.xticks(rotation=0)
plt.legend(title="Estado de diabetes", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Se observa una leve mayor proporción de diabetes en personas fumadoras en comparación con aquellas que no fuman. Esto sugiere que el tabaquismo podría estar asociado al estado de salud general, aunque no se presenta como un factor determinante por sí solo.

### 3.4 Relación entre edad y diabetes

```python
plt.figure(figsize=(7,5))

sns.boxplot(
    x="Estado_Diabetes_Descripcion",
    y="Edad",
    hue="Estado_Diabetes_Descripcion",
    data=df_final,
    palette="Purples",
    legend=False
)

plt.title("Relación entre edad y diabetes")
plt.xlabel("Estado de diabetes")
plt.ylabel("Edad")

plt.show()
```

Se observa que las personas con diabetes tienden a pertenecer a grupos etarios más altos, lo que sugiere una relación entre la edad y el desarrollo de la enfermedad.
